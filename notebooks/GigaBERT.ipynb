We load the model and tokenizer

```python
from transformers import BertTokenizer, BertForMaskedLM

tokenizer = BertTokenizer.from_pretrained("lanwuwei/GigaBERT-v4-Arabic-and-English")
model = BertForMaskedLM.from_pretrained("lanwuwei/GigaBERT-v4-Arabic-and-English")
```

We use the model for predicting masked tokens

```python
from transformers import pipeline

pipe = pipeline('fill-mask', model=model, tokenizer=tokenizer)

pipe("hey [MASK] are you doing")
```
