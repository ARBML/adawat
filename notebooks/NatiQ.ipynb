# NatiQ: Arabic Text to Speech


Get_models

```python
import requests
import json
URL = "https://tts.qcri.org/api/get_models"
headers={"Content-Type": "application/json;charset=utf-8"}
x = requests.post(URL, headers=headers)
y = json.loads(x.content)
print(y["model_list"][0]["id"]) # there are 3 models
print(y["model_list"][1]["id"]) 
print(y["model_list"][2]["id"])
```

Submit_job

```python
import requests
import json
URL = "https://tts.qcri.org/api/submit_job"
text=' اللغة العربية' 
model_id="21b418de-ab44-4bea-9fb1-7983ce81b070"
headers={"Content-Type": "application/json;charset=utf-8"}
params={"model_id":model_id, "text":text, "is_public":False}
x = requests.post(URL, headers=headers, data=json.dumps(params))
y = json.loads(x.content)
print(y["job_id"])
```

Get job_id.wav file

```python
import requests
import json
job_id=y["job_id"]
URL = "https://tts.qcri.org/api/audio/"+job_id+".wav"
headers={"Content-Type": "application/json;charset=utf-8"}

x = requests.get(URL, headers=headers)
from IPython.display import Audio
wn = Audio(URL, autoplay=True)
display(wn)
```
